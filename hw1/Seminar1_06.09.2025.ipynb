## 1. Pandas library <a class="anchor" id="chapter1"></a>

Pandas is a software library written for the Python programming language for data manipulation and analysis. In particular, it offers data structures and operations for manipulating numerical tables and time series. The name is derived from the term "panel data", an econometrics term for data sets that include observations over multiple time periods for the same individuals.

Pandas is mainly used for data analysis. It allows importing data from various file formats such as comma-separated values, SPSS and Stata files, Microsoft Excel and many others. Pandas allows various data manipulation operations such as merging, reshaping, selecting, as well as data cleaning, and data wrangling features.

Pandas is built on top of two core Python libraries—matplotlib for data visualization and NumPy for mathematical operations. Pandas acts as a wrapper over these libraries, allowing you to access many of matplotlib's and NumPy's methods with less code. For instance, pandas' .plot() combines multiple matplotlib methods into a single method, enabling you to plot a chart in a few lines.

[Cheat Sheet](https://pandas.pydata.org/Pandas_Cheat_Sheet.pdf)

The community agreed alias for pandas is pd, so loading pandas as pd is assumed standard practice for the pandas documentation.

```python
import pandas as pd
```

## 6. Grouping Statistics <a class="anchor" id="chapter6"></a>

A **groupby** operation involves some combination of splitting the object, applying a function, and combining the results. This can be used to group large amounts of data and compute operations on these groups.

```python
df
```

```python
def get_price_gr(x):
    if x < 45000:
        return 'low'
    elif x >=60000 and x<90000:
        return 'medium'
    elif x>=90000:
        return 'high'
```

```python
df['Price_groups'] = df['Price'].apply(get_price_gr)
# create a categorical variable based on a quantitative
df.head()
```

```python
df.Price_groups.value_counts()
```

```python
df.groupby('processor_brand')[['Price']].mean()  # mean Price grouped by Processor brand
```

```python
df.groupby('processor_brand')[['Price']].agg(['mean', 'std']) 
# mean and standard deviation of Price grouped by Processor brand
```

```python
df.groupby(['processor_brand', 'OS'])['Price'].mean().unstack() 
# mean Price grouped by Processor brand and Operating system 
```

Pivot table in pandas is an excellent tool to summarize one or more numeric variable based on two other categorical variables. It  has three main elements: 
1. <mark>index</mark> specifies the row-level grouping, 
2. <mark>columns</mark> specifies the column level grouping,
3. <mark>values</mark> are the numerical values that are summarized.

```python
df.pivot_table('Price', index='processor_brand', columns='primary_storage_type') 
# mean Price grouped by Processor brand and Primary storage type
```

```python
df.pivot_table('Price', index='processor_brand', columns='primary_storage_type', aggfunc='max')
# maximum price grouped by Processor brand and Primary storage type
```

## 9. Normality of the Distribution <a class="anchor" id="chapter9"></a>

```python
import warnings
warnings.filterwarnings('ignore')  # don't show warnings

sns.distplot(df.Price.dropna())  
# distribution plot
```

```python
from scipy.stats import norm  # normal distribution
sns.distplot(df.Price.dropna(), fit=norm)
```

```python
sns.distplot(df.Rating.dropna(), fit=norm)
```

```python
df.Price.skew()  # Skewness
```

The distribution is positively skewed (the frequent scores are clustered at the lower end and the tail points towards the higher or more positive scores)

```python
df.Price.kurtosis()  # Kurtosis
```

Platykurtic distribution is a statistical distribution characterized by its thinner tails and a flatter peak than the normal distribution.

```python
df.Rating.skew()  # Skewness
```

```python
df.Rating.kurtosis()  # Kurtosis
```

```python
# Красивые визуализации по данным df
sns.set_theme(style="whitegrid", context="talk", palette="Set2")

fig, axes = plt.subplots(1, 2, figsize=(16, 6), constrained_layout=True)

# 1) Распределение цен по OS (violin + точки)
sns.violinplot(data=df, x="OS", y="Price", inner="quartile", cut=0, ax=axes[0])
sample_n = min(300, len(df))
sns.stripplot(
    data=df.sample(n=sample_n, random_state=42),
    x="OS", y="Price",
    color="k", alpha=0.25, size=3, jitter=True, ax=axes[0]
)
axes[0].set_title("Распределение цены по OS")
axes[0].set_xlabel("")
axes[0].tick_params(axis='x', rotation=20)

# 2) Связь цены и рейтинга с трендом
sns.scatterplot(data=df, x="Price", y="Rating", hue="processor_brand", alpha=0.7, ax=axes[1])
sns.regplot(data=df, x="Price", y="Rating", scatter=False, lowess=True, color="black", ax=axes[1])
axes[1].set_title("Цена vs Рейтинг по бренду процессора")
axes[1].legend(title="processor_brand", bbox_to_anchor=(1.02, 1), loc="upper left", borderaxespad=0.)

sns.despine()
plt.show()
```
